#Desafio 5: calcular e analisar a correlação entre os g's e c's, expressões génicas e, os tipos celulares:


podemos pensar que os genes que têm o mesmo comportamento em todos os tipos celulares podem estar relacionados a funções vitais que toda célula precisa exercer. E agora é só tentar entender por que alguns genes são regulados de uma maneira em certos tipos celulares, mas em outros não?

Eu penso que os genes que têm alta correlação em todos os tipos celulares são aqueles relacionados à maquinaria da síntese de proteínas, pois toda célula precisa sintetizar proteínas para se manter viva (seria uma função vital, assim esses genes estariam ativos em todas essas células). Já aqueles genes que tem correlação com alguns tipos celulares, mas não tem com outros, podem ser genes que tem atividade específica, de acordo com o tipo celular eles estão ativos ou não.

```python
#armazenando os genes de g0 a g50 (51 genes)
G = dados.loc[:,'g0':'g50']
#armazenando os tipos celulares de c0 a c49 (51 tipos celulares)
C = dados.loc[:,'c0':'c50']

#contruinfo o dataset
#concatenação dos dados
# =  unir o conteúdo de duas strings unir genes com celulas
G_C = pd.concat([G,C], axis=1)
G_C.head()


```

```python
#gerando a matriz de correlação

corr = G_C.corr()
#filtrando a matriz de correlação para obter os dados de gene em um eixo 
#e os dados de célula em outro eixo, X g0 - g50, e y c0 - c50
corr_filtro = corr.loc[:'g50','c0':]

#plotando a matriz de correlação para os dados de viabilidade celular
#retira o fundo cinza
sns.set_theme(style="white")

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_filtro, dtype=bool), )

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 12))

# Generate a custom diverging colormap
# gerando cores diferentes
cmap = sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_filtro, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()

```
